### Q3. Use probability plots to determine which features are from normal distribution and which features are from uniform distribution. Write down your answers based on the generated figures. 
#### 3.1 Features from normal distribution: 
    Answer: x1, x2, x4, x5
#### 3.2 Features from uniform distribution: 
    Answer: x3

```python
f, axs = plt.subplots(nrows=2, ncols=5, figsize=(20, 8))

for i in range(1, 6):
    ax = axs[0, i-1]
    stats.probplot(df[f'x{i}'], dist='norm', plot=ax)
    ax.set_title(f'Normal Q-Q Plot x{i}')
    ax.set_xlabel("Normal Distribution")
    
    ax = axs[1, i-1]
    stats.probplot(df[f'x{i}'], dist='uniform', plot=ax)
    ax.set_title(f'Uniform Q-Q Plot x{i}')
    ax.set_xlabel("Uniform Distribution")

plt.tight_layout()
```

### Q4. Use scatterplot to explore the pair-wise relationships among these five features. Answer the following questions based the generated figures.
#### 4.1 Which feature pairs are positively correlated?
    Answer: pair (x2, x5)
#### 4.2 Which feature pairs are negatively correlated? 
    Answer (x2, x4), and (x4, x5).

```python
f, axs = plt.subplots(nrows=4, ncols=4, figsize=(15, 12))

for i in range(0, 4):
    for j in range(i+1, 5):
        ax = axs[i, j-1]
        ax.scatter(df[f'x{i+1}'], df[f'x{j+1}'])
        ax.set_title(f'Correlation between x{i} and x{j}')
        ax.set_xlabel(f'x{i}')
        ax.set_ylabel(f'x{j}')

plt.tight_layout()
```
